```python
dados_sudeste[meses_interessados]
```

```python
#Plotando o gráfico
color = ['c', 'g', 'y', 'r']
ax = dados_sudeste[meses_interessados].T.plot(figsize=(12,8), color=color)

ax.set_xticks(range(0, len(meses_interessados), 12))
ax.set_xticklabels(meses_interessados[::12])
# plt.ylim(0, 300000000)
plt.ylabel(ylabel="Gastos em milhões de reais", fontsize=14)
plt.title("Valores gastos no SUS por unidade federativa entre janeiro de 2011 até janeiro de 2016", fontsize=20)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left', fontsize=12)
plt.grid()
plt.show()
```

**Conclusões**

*   Os estados apresentam um comportamento similar de crescimento dos gastos ao longo do tempo.
*   O estado de São Paulo possui o dobro de gastos em relação ao segundo colocado Minas Gerais. Esse valor pode ser associado ao fato de que a população de São Paulo (46 milhões aproximadamente) é 2.2 vezes maior que a população de Minas Gerais (21 milhões aproximadamente).
*   Geralmente os estados selecionados apresentam uma tendência de crescimento de gastos no inícios dos anos, o que pode ser atribuído em parte pelos surtos de dengue.
*   Apesar do Rio de Janeiro ter uma população maior que o Paraná (cerca de 6 milhões de pessoas a mais), possui menores gastos no SUS.
*   Apesar do Rio de Janeiro ter uma população maior que a Bahia (cerca de 3 milhões de pessoas a mais), possui um invetimento equivalente ao longo do tempo.
